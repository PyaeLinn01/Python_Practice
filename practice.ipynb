## Append in List for Isomorphic

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        s = list(s)
        t = list(t)

        s_indices = []
        t_indices = []

        for ss in s:
            s_indices.append(s.index(ss))
        
        for tt in t:
            t_indices.append(t.index(tt))

        return s_indices == t_indices

              
Solution().isIsomorphic("egg", "add")            

```

## From list [9,9] to 99

```python
lst = [9,9]

result = 0
for digit in lst:
    #print(digit)
    result = result * 10 + digit

print(result)
```

## [9,9] to [1, 0, 0]

```python
digits = [9,9]
to_sum = 0
for d in digits:
    #print(d)
    
    to_sum = to_sum * 10 + d
to_sum += 1 #100
print(to_sum)
result = []
for s in str(to_sum):
    #print(s)
    
    result.append(int(s))
print(result)

print(type(result))    
```
